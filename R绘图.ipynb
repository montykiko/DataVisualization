## title

## size

```r
ggplot(mtcars,aes(mpg,drat,colour = factor(carb))) + geom_point(aes(size = 4))
```

## annotation

```r
p <-ggplot(mtcars,aes(mpg,drat,colour = factor(carb))) + geom_point(aes(size = gear))
```

```r
p + annotate("text",x=18,y=4,label = "Green Area")
```

## Adding line

```r
p + annotate("text",x=18,y=4,label = "Green Area") + geom_hline(yintercept = 4) + geom_vline(xintercept = 18)
```
